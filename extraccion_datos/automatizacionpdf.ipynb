<h3>Repositorio de PDFs</h3>

Library Of Congress Web Archiving Program. (2019) .gov PDF dataset. [Washington, D.C.: Library of Congress Web Archiving Program] [Software, E-Resource] Disponible en https://www.loc.gov/item/2020445568/.

<h3>Listado de archivos</h3>

```python
archivos = os.listdir("repositorioPDFs")
archivos
```

###### Expresión regular:

<img src="expresion_regular_email.jpg" width="400" align="left"> <br><br><br>
<br><h3>Implementación con PyMuPDF</h3>

```python
correos = []
for archivo in archivos: 
    documento = pymupdf.open("repositorioPDFs/" + archivo)
    contenido = ""
    for pagina in documento:
        contenido += pagina.get_text()
    correos.extend(re.findall("[a-z0-9\-_\.]+@[a-z0-9\-_\.]+\.[a-z]+", 
                              contenido, 
                              re.IGNORECASE))
correos
```

<h3>Implementación con PyPDF</h3>

```python
correos = []
for archivo in archivos: 
    documento = PdfReader("repositorioPDFs/" + archivo)
    contenido = ""
    for pagina in documento.pages:
        contenido += pagina.extract_text()
    correos.extend(re.findall("[a-z0-9\-_\.]+@[a-z0-9\-_\.]+\.[a-z]+", 
                              contenido, 
                              re.IGNORECASE))
correos
```
