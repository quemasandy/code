# Ilustrando la Aumentación de Datos con Imágenes
### Librerías e imagen de ejemplo

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import PIL

ave = PIL.Image.open("ave.jpg")
plt.imshow(ave)
plt.show()
```

# Más datos con diferentes niveles de saturación

```python
fig = plt.figure(figsize = (9, 3))

for nivel in range(0, 10):
  plt.subplot(1, 10, nivel + 1)
  plt.imshow(tf.image.adjust_saturation(ave, nivel + 1))
  plt.axis("off")

plt.show()
```

# Más datos con diferentes niveles de brillo

```python
fig = plt.figure(figsize = (9, 3))

for nivel in range(0, 10):
  plt.subplot(1, 10, nivel + 1)
  plt.imshow(tf.image.adjust_brightness(ave, (nivel + 1) / 10))
  plt.axis("off")

plt.show()
```

# Más datos con recortes aleatorios

```python
fig = plt.figure(figsize = (9, 3))

for nivel in range(0, 10):
  plt.subplot(1, 10, nivel + 1)
  plt.imshow(tf.image.random_crop(ave, size = [400, 400, 3]))
  plt.axis("off")

plt.show()
```

# Más datos con rotaciones

```python
fig = plt.figure(figsize = (11, 3))

for grados in range(0, 360, 45):
  plt.subplot(1, 10, grados//45 + 1)
  plt.imshow(ave.rotate(grados, fillcolor = "darkgreen"))
  plt.axis("off")

# tf.image.flip_left_right(ave)
# tf.image.flip_up_down(ave)

plt.show()
```

# Más datos con escala de grises

```python
fig = plt.figure(figsize = (8, 3))

plt.subplot(1, 2, 1)
plt.imshow(ave)

plt.subplot(1, 2, 2)
plt.imshow(tf.image.rgb_to_grayscale(ave), cmap="gray")

plt.show()
```
