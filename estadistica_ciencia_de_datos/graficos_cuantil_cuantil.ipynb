<h1 style="color:purple">Gráfico cuantil-cuantil (Q-Q plot)</h1>
<h2 style="color:pink">Datos sintéticos: altura de personas</h2>

```python
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.api import qqplot
from scipy.stats import norm

# quitar notación científica
np.set_printoptions(suppress = True)

# altura promedio
altura_promedio = 170 

# desviación estándar de la altura
dev_std_altura = 8 

# semilla para reproducibilidad
np.random.seed(1)

# tamaño muestra
n = 100

# datos sintéticos
alturas = np.random.normal(altura_promedio, dev_std_altura, n)

# Creación de histograma para las alturas 
plt.hist(alturas, bins=40, density=True, color="pink", alpha=0.7)
plt.box(False)
plt.show()
```

<h1 style="color:purple">Gráfico cuantil-cuantil (Q-Q plot)</h1>
<h2 style="color:pink">Paso a paso y vía statsmodels</h2>

```python
# Cálculo manual paso a paso

# Paso 1: Ordenar datos
alturas.sort()
# Paso 2: Crear ranking
rank = np.arange(1, len(alturas) + 1)
# Paso 3: Calcular cuantiles y z valores
cuantiles = (rank - 0.5) / rank.size
z_scores = norm.ppf(cuantiles)
# Paso 4: Graficar 
plt.scatter(z_scores, alturas)
plt.show()

# Cálculo vía statsmodels

qqplot(alturas, line="s") 
plt.show()
```
