# Asimetría (o sesgo) de un Conjunto de Datos
## Datos sintéticos con diferentes asimetrías

```python
####### Asimétrica positivamente #######
# Interpretación: La mayoría de los clientes hacen compras pequeñas
# y solo algunos hacen compras grandes.
import numpy as np
from scipy.stats import skew
import matplotlib.pyplot as plt
import seaborn as sns

monto_max = 600
monto_min = 1

for paso in range(22):
  # Datos sintéticos aleatorios diseñados experimentalmente para el ejemplo
  # n = 3000
  datos = np.random.normal(89 + paso*12, 21 + paso*3, 1900)
  datos[datos > monto_max] = monto_max
  datos[datos < monto_min] = monto_min
  montos_compras = np.concatenate([datos,
                                   np.random.uniform(monto_min + 99, monto_max,
                                                     1100)])
  # Calcular asimetría
  # Coeficiente Fisher-Pearson de asimetría,
  #asimetria = np.mean((montos_compras-np.mean(montos_compras))**3) / (np.std(montos_compras)**3)
  asimetria = skew(montos_compras)

  # Graficar la distribución
  plt.figure(figsize=(6, 5))
  sns.histplot(montos_compras, bins=100, color="purple", edgecolor="purple")
  plt.yticks(np.arange(0, 221, 20))
  plt.xticks(np.arange(0, 601, 50))
  plt.xlabel(f"Monto de Compra (Asimetría: {asimetria:.3f})", fontsize=12)
  plt.ylabel("Frecuencia", fontsize=12)
  plt.box(False)
  plt.show()

####### Asimétrica negativamente #######
# Interpretación: La mayoría de los clientes hacen compras grandes
# y solo algunos hacen compras pequeñas.

for paso in range(22):
  # Datos sintéticos aleatorios diseñados experimentalmente para el ejemplo
  # n = 3000
  datos = np.random.normal(270 + paso*12, 87 - paso*3, 1900)
  datos[datos > monto_max] = monto_max
  datos[datos < monto_min] = monto_min
  montos_compras = np.concatenate([datos,
                                   np.random.uniform(monto_min, monto_max-100,
                                                     1100)])
  # Calcular asimetría
  asimetria = skew(montos_compras)

  # Graficar la distribución
  plt.figure(figsize=(6, 5))
  sns.histplot(montos_compras, bins=100, color="purple", edgecolor="purple")
  plt.yticks(np.arange(0, 221, 20))
  plt.xticks(np.arange(0, 601, 50))
  plt.xlabel(f"Monto de Compra (Asimetría: {asimetria:.3f})", fontsize=12)
  plt.ylabel("Frecuencia", fontsize=12)
  plt.box(False)
  plt.show()
```
