# Ilustración de resultados estadísticamente significativos

```python
import numpy as np
from scipy import stats

# Semilla para reproducibilidad
np.random.seed(8)

# Tamaños de las muestras de alturas
tamanios = [1, 2, 10, 30, 100]

for t in range(len(tamanios)):
  print("TAMAÑO DE LA MUESTRA:", tamanios[t], "\n")
  if t == 0:
    # Generando la primera muestra
    americanos = np.random.normal(173, 5, tamanios[t])
    europeos = np.random.normal(178, 5, tamanios[t])
  else:
    # incrementando el tamaño de las muestras de alturas
    mas_americanos = np.random.normal(173, 5, tamanios[t]-tamanios[t-1])
    americanos = np.concatenate((americanos, mas_americanos))
    mas_europeos = np.random.normal(178, 5, tamanios[t]-tamanios[t-1])
    europeos = np.concatenate((europeos, mas_europeos))

  # Redondeo para propósitos de visualización
  americanos = np.round(americanos, 2)
  europeos = np.round(europeos, 2)

  # Alturas promedio
  print("Muestra de alturas de Americanos:", americanos)
  print("Altura promedio de Americanos:", americanos.mean())
  print("\nMuestra de alturas de Europeos:", europeos)
  print("Altura promedio de Europeos:", europeos.mean())

  # Prueba t de Student - muestras independientes de igual tamaño y varianzas iguales
  t_stat, p_valor = stats.ttest_ind(americanos, europeos)

  # Interpretación de los resultados:
  alfa = 0.05
  if p_valor < alfa:
    print("\nSe rechaza la hipótesis nula. Hay evidencia apoyando que \
    las alturas medias de los Americanos y de los Europeos son diferentes.")
  else:
    print("\nNo se puede rechazar la hipótesis nula. Hay poca evidencia \
    respaldando la hipótesis alternativa de que las alturas medias \
    de los Americanos y los Europeos son diferentes.")

  print("\n( Estadístico t =", t_stat, "Valor P =", p_valor, ")"+"\n"*4)
```
