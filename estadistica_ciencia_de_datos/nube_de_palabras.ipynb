# <b>Análisis de Texto</b>: Nube de Palabras

# Paquetes

```python
from wordcloud import WordCloud
import nltk
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import requests
```

# Texto descargado de "Proyecto Gutenberg"

```python
url_libro = "https://www.gutenberg.org/cache/epub/56454/pg56454.txt"
libro = requests.get(url_libro).text
print(libro[2000:2340])
```

# Nube de Palabras

```python
nube = WordCloud().generate(libro)
plt.imshow(nube)
plt.show()
```

# Palabras vacías (stop words)

```python
nltk.download("stopwords")
palabras_irrelevantes = nltk.corpus.stopwords.words("spanish")

# Agregar más palabras irrelevantes en función del contexto
palabras_irrelevantes.append("á")
palabras_irrelevantes
```

# Nube de Palabras quitando Palabras Vacías

```python
nube = WordCloud(stopwords = palabras_irrelevantes).generate(libro)
plt.imshow(nube)
plt.axis("off")
plt.show()
```

# Nube de Palabras Personalizada

```python
nube = WordCloud(max_words = 50,
                 colormap = "magma",
                 background_color = "lavenderblush",
                 width = 700,
                 height = 350,
                 stopwords = palabras_irrelevantes
                 ).generate(libro)
plt.imshow(nube)
plt.axis("off")
plt.show()
```

# Nubes de Palabras con Mascaras

```python
# Cargar figuras / mascaras
mascara1 = np.array(Image.open("estrella1.png"))
mascara2 = np.array(Image.open("estrella2.png"))

# Ajusta el mapa de color
plt.set_cmap("Greys")

# Figura con dos subfiguras
fig, axes = plt.subplots(1, 2, figsize=(5, 10))

# Incluir la primera mascara
axes[0].imshow(mascara1)
axes[0].axis("off")
# Incluir la segunda mascara
axes[1].imshow(mascara2)
axes[1].axis("off")

plt.show()
```

# Nube de Palabras con Mascara versión 1

```python
nube = WordCloud(mask = mascara1,
                 contour_width = 3,
                 contour_color = "pink",
                 max_words = 50,
                 colormap = "magma",
                 background_color = "lavenderblush",
                 stopwords = palabras_irrelevantes
                 ).generate(libro)
plt.imshow(nube)
plt.axis("off")
plt.show()
```

# Nube de Palabras con Mascara versión 2

```python
nube = WordCloud(mask = mascara2,
                 contour_width = 5,
                 contour_color = "pink",
                 stopwords = palabras_irrelevantes,
                 max_words = 50,
                 background_color = "lavenderblush",
                 colormap = "magma",
                 ).generate(libro)
plt.imshow(nube)
plt.axis("off")
plt.show()
```
