# Ejemplos de la Distribución Normal Estándar

```python
import matplotlib.pyplot as plt
import numpy as np
import math

tamano = 1000000
compartimentos = 1000

media_surenos = 170
desv_surenos = 8

media_nortenos = 180
desv_nortenos = 16


fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(8, 2.5), dpi=600)
plt.subplots_adjust(hspace=0.5)


surenos = np.random.normal(media_surenos, 
                           desv_surenos,
                           tamano)

nortenos = np.random.normal(media_nortenos,
                            desv_nortenos, 
                            tamano)

def subfigura(subfig, titulo, datos, media, desv_est, color):
    subfig.set_title(titulo, fontsize=10)
    subfig.hist(datos, bins=compartimentos, color=color, alpha=0.5)

    subfig.axvline(media - 3*desv_est, color=color, alpha=0.2)
    subfig.axvline(media - 2*desv_est, color=color, alpha=0.2)
    subfig.axvline(media - 1*desv_est, color=color, alpha=0.2)
    
    subfig.axvline(media, color=color, alpha=0.2)

    subfig.axvline(media + 1*desv_est, color=color, alpha=0.2)
    subfig.axvline(media + 2*desv_est, color=color, alpha=0.2)
    subfig.axvline(media + 3*desv_est, color=color, alpha=0.2)
    subfig.get_yaxis().set_visible(False)
    subfig.set_xticks((round(media - 3*desv_est, 0),
                       round(media - 2*desv_est, 0),
                       round(media - 1*desv_est, 0),
                       round(media, 0),
                       round(media + 1*desv_est, 0),
                       round(media + 2*desv_est, 0),
                       round(media + 3*desv_est, 0)))
    subfig.tick_params(labelsize=5)
    

    
subfigura(ax1, "Sureños", surenos, surenos.mean(), surenos.std(), "purple")
subfigura(ax2, "Norteños", nortenos, nortenos.mean(), nortenos.std(), "steelblue")

ax3.set_title("Ambos", fontsize=10)
ax3.hist(surenos, bins=compartimentos, alpha=0.5, color="purple")
ax3.hist(nortenos, bins=compartimentos, alpha=0.5, color="steelblue")
ax3.set_xticks((110, 134, 156, 180, 204, 226, 250))
ax3.tick_params(labelsize=5)
ax3.get_yaxis().set_visible(False)


z_surenos = (surenos - media_surenos) / desv_surenos
subfigura(ax4, "Sureños Estandarizados", z_surenos, z_surenos.mean(), z_surenos.std(), "purple")

z_nortenos = (nortenos - media_nortenos) / desv_nortenos
subfigura(ax5, "Norteños Estandarizados", z_nortenos, z_nortenos.mean(), z_nortenos.std(), "steelblue")

ax6.set_title("Ambos Estandarizados", fontsize=10)
ax6.hist(z_surenos, bins=compartimentos, alpha=0.5, color="purple")
ax6.hist(z_nortenos, bins=compartimentos, alpha=0.5, color="steelblue")
ax6.set_xticks((-3, -2, -1, 0, 1, 2, 3))
ax6.tick_params(labelsize=5)
ax6.get_yaxis().set_visible(False)

plt.show()
```
